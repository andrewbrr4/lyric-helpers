```python
from word_associations import associate
```

## Explore

Adjust `surprise` from 1 (near-synonyms) to 10 (unexpected connections).

```python
parts_of_speech = ["noun", "verb", "adjective", "adverb"]

for pos in parts_of_speech:
    results = associate("charcoal", surprise=4, pos=pos)
    words = [result[0] for result in results]
    print(f"{pos.capitalize()}: {', '.join(words)}")
```
